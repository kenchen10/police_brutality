# Datasets
1. **pb_data**: police brutality data
2. **median_income**: US Census data containing the median household income in all US cities. Features: "Geographic area": State "City": City "Median Income": Median household income in given city.
3. **complete_hs**: US Census data containing the high school graduation rate for people over 25 in all US cities. Features: "Geographic area": State "City": City "percent_completed_hs": Percentage of given city's population above 25 that has graduated high school.
4. **police_killings**: US Census data containing the median household income in all US cities. Features: "Geographic area": State "City": City "Median Income": Median household income in given city.
5. **race**: US Census data regarding racial demographic in all US cities. Features: "Geographic area": State "City": City "share_white": Percentage of the city's population that is white "share_black": Percentage of the city's population that is black "share_native_american": Percentage of the city's population that is native american "share_asian": Percentage of the city's population that is asian "share_hispanic": Percentage of the city's population that is hispanic

# Median Income

```python
median_income.head()
```

```python
median_income = median_income[median_income["Median Income"] != "(X)"]
median_income = median_income[median_income["Median Income"] != "-"]
median_income = median_income[median_income["Median Income"] != "2,500-"]
median_income = median_income.dropna()
```

```python
median_income.groupby("Median Income").size().head()
```

```python
median_income["Median Income"] = median_income["Median Income"].replace({"2,500-":"2500", "250,000+": "250000"})
median_income['Median Income'] = pd.to_numeric(median_income['Median Income'], errors='coerce')
```

```python
np.average(median_income["Median Income"])
```

```python
police_killings["race"] = police_killings["race"].replace({"A":"asian", "B": "black", "W": "white", "H": "hispanic", "O": "other", "N": "native american"})
```

```python
race = police_killings.groupby("race").size().sort_values()
labels = race.index
vals = race.values
race.plot.bar()
plt.ylabel("Number of Deaths")
```
